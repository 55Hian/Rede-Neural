```python
import numpy as np
import math 

entradas =  [       [0, 0], 
                    [0, 1],
                    [1, 0],
                    [1, 1]
            ]

resultadoEsperado = [0,
                     0, 
                     0, 
                     1]

saida = []

erro = []

errot = 1

soma = 0
pesos = 0
taxaAprendizado = 0.1

i = 0
coletaIndice = 0
trava = 0

while saida != resultadoEsperado:
    for x in entradas:
        soma = ((x[0] + x[1]) * pesos)

        if soma >= 1:
            saida.append(1)
        else:
            saida.append(0)

    print(saida)
    print(resultadoEsperado)

    for x in resultadoEsperado:
        erro.append(x - saida[i])
        i += 1

    i = 0

    print(erro)

    for x in erro:
        if x != 0 and trava == 0:
            coletaIndice = i
            print("indice do erro")
            print(coletaIndice)
            print("valor do numero errado")
            print(erro[coletaIndice])
            trava = 1
            break
        i += 1


    if saida != resultadoEsperado:
        saida = []
        erro = []

    # Adicionando verificações de limite
    if coletaIndice < len(entradas) and coletaIndice < len(erro):
        pesos = pesos + (taxaAprendizado * entradas[coletaIndice][1] * erro[coletaIndice])
    else:
        print("Índice fora do alcance.")

    print("saiu do for")

print("cabou")
```
